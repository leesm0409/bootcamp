```python
#1월과 8월의 최고 기온의 빈도수를 한번에 표현하기
import csv
import matplotlib.pyplot as plt

# 01월, 08월 최고 기온 정보를 담는 리스트 선언
jan_list=[]
aug_list=[]

# 1) 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# 2) csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#헤더(데이터 정보) 저장하기
header = next(data)
print(header)


#2) 기온 데이터 출력 -> 날짜와 최고 기온만 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> result 리스트에 담지 않기
  if row[-1] != '': #결측치 데이터 처리
    #01월과 08월 데이터를 담기
    month = row[0].split('-')[1]
    if month == '08':
      aug_list.append(float(row[-1]))

    if month == '01':
      jan_list.append(float(row[-1]))


f.close()

#5)최고 기온 데이터 시각화 -> 히스토그램 , 빈도수 지정 = bins
plt.figure(dpi=100)

plt.title("서울의 8월 최고기온 히스토그램")
plt.hist(jan_list, bins = 100, color = 'b', label='1월의 최고 기온')
plt.hist(aug_list, bins = 100, color = 'r', label = '8월의 최고 기온')
plt.legend()
plt.show()
```

```python
#최저 기온 파랑, 평균 초록, 최고 기온 빨강으로 표시
#벌례를 적용하여 하나의 히스토그램으로 그리기

import csv
import matplotlib.pyplot as plt

# 원하는 월을 입력 받기
choice_month= input("원하는 월을 입력하세요 :")
low_temp = []
avg_temp = []
high_temp = []

# 1) 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# 2) csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#헤더(데이터 정보) 저장하기
header = next(data)
print(header)


#2) 기온 데이터 출력 -> 날짜와 최고 기온만 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> result 리스트에 담지 않기
  if row[-1] != '': #결측치 데이터 처리
    #01월과 08월 데이터를 담기
    month = row[0].split('-')[1]
    if month == choice_month:
      low_temp.append(float(row[-2]))
      high_temp.append(float(row[-1]))
      avg_temp.append(float(row[-3]))




f.close()

#5)최고 기온 데이터 시각화 -> 히스토그램 , 빈도수 지정 = bins
plt.figure(dpi=100)

plt.title("서울의 8월 최고기온 히스토그램")
plt.hist(low_temp, bins = 100, color = 'b', label= ('최저기온'))
plt.hist(high_temp, bins = 100, color = 'r', label = '8월의 최고 기온')

plt.legend()
plt.show()
```

```python
#최저 기온 파랑, 평균 초록, 최고 기온 빨강으로 표시
#벌례를 적용하여 하나의 히스토그램으로 그리기

import csv
import matplotlib.pyplot as plt

# 원하는 월을 입력 받기, 얻고자 하는 결론에 대한 변수설정
choice_month= input("원하는 월을 입력하세요 :")

if len(choice_month) ==1:
  choice_month = '0' + choice_month

low_temp = []
avg_temp = []
high_temp = []

# 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#헤더(데이터 정보) 저장하기
header = next(data)
print(header)


#2) 기온 데이터 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> result 리스트에 담지 않기
  if row[-1] and row[-2] and row [-3] != '': #결측치 데이터 처리 (-1,-2,-3 모두 해결해줘야 됨 )
    #데이터를 담기
    month = row[0].split('-')[1]
    if month == choice_month:
      low_temp.append(float(row[-2]))
      high_temp.append(float(row[-1]))
      avg_temp.append(float(row[-3]))


f.close()

#5)최고 기온 데이터 시각화 -> 히스토그램 , 빈도수 지정 = bins
plt.figure(dpi=100)

plt.title("서울의 " + choice_month + "월 최고,평균,최저 기온 히스토그램")

plt.hist(high_temp, bins = 100, color = 'r', label = choice_month + '월 최고 기온')
plt.hist(avg_temp, bins = 100, color = 'g', label = choice_month + '월 평균 기온')
plt.hist(low_temp, bins = 100, color = 'b', label= choice_month + '월 최저기온')
plt.legend()
plt.show()
```
