기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제고 몇도 였을까?

```python
import csv

max_temp = -999  # 최고 기온 값을 저장할 변수 -> 명확한 값을 위해 극적으로 초기값 설정
max_date = '' #최고 기온이 가장 높았던 날짜를 저장 할 변수

# 1) 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# 2) csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#next()함수는 한줄 읽고, 커서 (읽어드릴 위치) 한칸이동
#헤더(데이터 정보) 저장하기
header = next(data)

#2) 기온 데이터 출력 -> 날짜와 최고 기온만 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> 명확한 결과를 위해 극적인 값으로 초기화
  if row[-1] =='':
    row[-1] = -999

# 2-2) 최고기온 실수형으로 형변환
  row[-1] = float(row[-1])

# 3) 최고 기온과 날짜 찾기
  if max_temp < row[-1]:
    max_temp = row[-1] #최고기온
    max_date = row[0] #최고 기온의 날짜


# 4) 파일을 닫는다
f.close()


#5) 결과 출력
print('기상 관측 이래, 서울의 최고 기온이 높았던 날은 %s로 %.2f 였습니다.' %(max_date,max_temp))
```

기상 관측 이래, 서울의 최저 기온이 가장 낮았던 날은 언제고 몇도 였을까?

```python
import csv

low_temp = 999  # 최고 기온 값을 저장할 변수 -> 명확한 값을 위해 극적으로 초기값 설정
low_date = '' #최고 기온이 가장 높았던 날짜를 저장 할 변수

# 1) 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# 2) csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#next()함수는 한줄 읽고, 커서 (읽어드릴 위치) 한칸이동
#헤더(데이터 정보) 저장하기
header = next(data)

#2) 기온 데이터 출력 -> 날짜와 최고 기온만 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> 명확한 결과를 위해 극적인 값으로 초기화
  if row[-2] =='':
    row[-2] = 999

# 2-2) 최고기온 실수형으로 형변환
  row[-2] = float(row[-2])

# 3) 최고 기온과 날짜 찾기
  if low_temp > row[-2]:
    low_temp = row[-2] #최저기온
    low_date = row[0] #최고 기온의 날짜


# 4) 파일을 닫는다
f.close()


#5) 결과 출력
print('기상 관측 이래, 서울의 최저 기온이 낮았던 날은 %s로 %.2f 였습니다.' %(low_date,low_temp))
```

기온데이터 시각화 하기

```python
#관측이래, 서울의 최고 기온의 데이터를 시각화
import csv
import matplotlib.pyplot as plt

# 최고 기온 정보를 담는 리스트 선언
result=[]

# 1) 파일을 연다.
f = open('/content/seoul_2023.csv','r', encoding = 'euc-kr')

# 2) csv 파일을 읽는다
data = csv.reader(f)
# , 로 되어있으면 delimiter = ','생략 가능

#헤더(데이터 정보) 저장하기
header = next(data)
print(header)


#2) 기온 데이터 출력 -> 날짜와 최고 기온만 출력
for row in data:
  #2-1) 결측치 데이터 처리 -> result 리스트에 담지 않기
  if row[-1] != '': #결측치가 아닌 최기 기온데이처 실수형 변환하여 result 리스트에 담기
    result.append(float(row[-1]))


f.close()

#5)최고 기온 데이터 시각화
plt.figure(dpi=200)
plt.plot(result,'r')
plt.show()
```
