## 1.이진 분류 데이터셋(forge) 확인하기

```python
import mglearn
import matplotlib.pyplot as plt
import koreanize_matplotlib

import warnings
warnings.filterwarnings('ignore')

#데이터셋 다운로드
X,y = mglearn.datasets.make_forge()
```

## 데이터 확인하기

```python
print('X.shape: ', X.shape)
print('y.shape: ', y.shape)
```

```python
plt.figure(dpi=100)

# 산점도 그리기
mglearn.discrete_scatter(X[:,0],X[:,1],y)

plt.legend(['클래스0','클래스1'],loc=4)
plt.xlabel('첫번째 특성')
plt.ylabel('두번째 특성')
plt.show()
```

## 2. 회귀 데이터셋(wave)확인하기

```python
X,y = mglearn.datasets.make_wave(n_samples=40)
```

```python
print('X.shape: ', X.shape)
print('y.shape: ', y.shape)
```

```python
 #산점도 => 연속적인 값 그릴때 이용
 plt.figure(dpi=100)

# 산점도 그리기 -> x와 y의 연관성을 위한
plt.plot(X,y,'o')
#plt.ylim(-3,3)

plt.xlabel('특성')
plt.ylabel('타깃')
plt.show()
```

## 3. 위스콘신 유방암 데이터셋

*   이진분류 문제로 정의
*   0 = malignant, 1 = benign



```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer['target'] #0 = malignant, 1 = benign
```

### 데이터 확인하기

```python
print("cancer['data'].shape : ", cancer['data'].shape ) #569건의 데이터와 30개의 특성으로 이루어진 데이터셋
print("cancer['target'].shape : ", cancer['target'].shape) #0 = malignant, 1 = benign
```

```python
cancer.feature_names
```

## 데이터와 타깃의 연관성 확인 -> 데이터분석

```python
import pandas as pd

# 데이터프레임을 사용하여 데이터 분석 -> data와 target의 연관성 분석
cancer_df = pd.DataFrame(cancer['data'],columns= cancer.feature_names)
cancer_df.head()
```

```python
# 각 feature 들의 산점도 행렬 30*30
pd.plotting.scatter_matrix(cancer_df, c = cancer['target'], figsize = (15,15),
                           marker = 'o', alpha=.8)

plt.show()
```

```python
import numpy as np

plt.imshow([np.unique(cancer['target'])])
_ = plt.xticks(ticks=np.unique(cancer['target']),labels=cancer['target_names'])
```

## 4. 캘리포니아 주택 가격 데이터셋



*   1990년 캘리포니아 평균 주택가격
*   주택가격을 예측하는 회기 문제로 정의



```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

### 데이터 확인하기

```python
print("housing['data'].shape : ", housing['data'].shape ) #20640건의 데이터와 8개의 특성으로 이루어진 데이터셋
print("housing['target'].shape : ", housing['target'].shape) #1990년 캘리포니아의 평균 집값
```

```python
print(housing.DESCR)
```

```python
housing['target']
```

### 데이터와 타깃의 연관성 확인 -> 데이터분석

```python
import pandas as pd

# 데이터프레임을 사용하여 데이터 분석 -> data와 target의 연관성 분석
housing_df = pd.DataFrame(housing['data'],columns= housing.feature_names)
housing_df.head()
```

```python
#타겟값 컬럼으로 추가
housing_df['target'] = housing['target']
housing_df.info()
```

```python
housing_df['target'].max()
```

```python
#캘리포니아에서 평균 집값이 housing_df['target'].max()와 같은 애들만 지도에 표시

import folium

my_map = folium.Map(location=[housing_df['Latitude'].mean(),housing_df['Longitude'].mean()])

for n in housing_df.index:
  price = (housing_df.loc[n,'target'])*100000

  if housing_df.loc[n, 'target'] >= housing_df['target'].max():
    folium.Marker([housing_df.loc[n, 'Latitude'], housing_df.loc[n, 'Longitude']],
                  popup=str(price)+'$',
                  icon=folium.Icon(color='red', icon='fa-home',prefix='fa')).add_to(my_map)

my_map
```
